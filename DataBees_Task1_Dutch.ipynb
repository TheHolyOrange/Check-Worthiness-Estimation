# PRE-PROCESSING OF DATA

```python
def tolower(text):
	return text.lower()
```

```python
def removepunctuation(text):
	clean="".join([i for i in text if i not in string.punctuation])
	return clean
```

```python
nltk.download('stopwords')
stopwords_list = stopwords.words('spanish')
stopwords_list
```

```python
def removestopwords(text):
	clean = " ".join([word for word in text.split() if word not in stopwords_list])
	return clean
```

```python
def deEmojify(text):
    regrex_pattern = re.compile(pattern = "["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           "]+", flags = re.UNICODE)
    return regrex_pattern.sub(r'',text)
```

```python
df_train['class_label'] = df_train['class_label'].replace(['Yes','No'],[1,0])
df_test['class_label'] = df_test['class_label'].replace(['Yes','No'],[1,0])
def preprocess(df):
  df['tweet_text']= df['tweet_text'].apply(lambda x:tolower(x))
  df['tweet_text']= df['tweet_text'].apply(lambda x:removepunctuation(x))
  df['tweet_text']= df['tweet_text'].apply(lambda x:removestopwords(x))
  df['tweet_text']= df['tweet_text'].apply(lambda x:deEmojify(x))
  return df
```

```python
#Converting the Text to a string and class_label to integer
df_train.tweet_text=df_train.tweet_text.astype(str)
df_train.class_label=df_train.class_label.astype(int)
#Applying the function
train = preprocess(df_train)
train_data=train[["tweet_text","class_label"]]
train_data
```

```python
#Converting the Text to a string and class_label to integer
df_test.tweet_text=df_test.tweet_text.astype(str)
df_test.class_label=df_test.class_label.astype(int)
#Applying the function
test = preprocess(df_test)
test_data=test[["tweet_text","class_label"]]
test_data
```

# APPLYING TRANSFORMER MODELS

```python
def predict_result(bert,df):
    predictions, raw_outputs = bert.predict(df['tweet_text'].tolist())

    return predictions
```

```python
def find_accuracy(bert,df):
    result, model_outputs, wrong_predictions = bert.eval_model(df)

    predictions, raw_outputs = bert.predict(df['tweet_text'].tolist())
    print("predictions:")
    print(predictions)

    bads = {}
    for pred in wrong_predictions:
        if pred.label in bads:
            bads[pred.label] += 1
        else:
            bads[pred.label] = 1
    print("wrong predictions:")
    wrong = len(wrong_predictions)
    total = len(df)
    print(str(wrong) + ' wrong out of ' + str(total))
    print(bads)
    accuracy = (total-wrong)/total*100
    print("Accuracy: "+str(accuracy)+"%")

    print("Classification report: ")
    from sklearn.metrics import classification_report
    print(classification_report(df.class_label, predictions))
```

# TRADITIONAL METHODS

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

texts_train = train_data["tweet_text"]
labels_train = train_data["class_label"]
texts_test = test_data["tweet_text"]
labels_test = test_data["class_label"]

# Convert the texts into feature vectors
vectorizer = CountVectorizer()
X_train = vectorizer.fit_transform(texts_train)
X_test = vectorizer.transform(texts_test)
```

```python
model = MultinomialNB()
model.fit(X_train, labels_train)
```

```python
labels_pred = model.predict(X_test)
```

```python
report = classification_report(labels_test, labels_pred)
print(report)
```

```python
model = SVC(kernel='linear')
model.fit(X_train, labels_train)
```

```python
labels_pred = model.predict(X_test)
```

```python
report = classification_report(labels_test, labels_pred)
print(report)
```

```python
model = LogisticRegression()
model.fit(X_train, labels_train)
```

```python
labels_pred = model.predict(X_test)
```

```python
report = classification_report(labels_test, labels_pred)
print(report)
```
