# PRE-PROCESSSING OF DATA

Conversion to Lowercase

```python
def tolower(text):
	return text.lower()
```

Removing Punctuation

```python
def removepunctuation(text):
	clean="".join([i for i in text if i not in string.punctuation])
	return clean
```

**Stemming**

Stemming is a method of normalization of words in NLP.

It is a technique in which a set of words in a sentence are converted into a sequence to shorten its lookup. In this method, the words having the same meaning, but with some variations according to the context or sentence, are normalized.

In other words, there is one root word, but there are many variations of the same words. For example, the root word is “eat” and it’s variations are “eats, eating, eaten” and so on.

In this way, with the help of Stemming in Python, we can find the root word of any variations.

```python
from nltk.stem import PorterStemmer
e_words= ["wait", "waiting", "waited", "waits"]
ps = PorterStemmer()
for w in e_words:
	rootWord = ps.stem(w)
	print("Stemming for {} is {}".format(w, rootWord))
```

**Lemmatisation**

Lemmatisation in NLTK is the algorithmic process of finding the lemma of a word depending on its meaning and context.
It helps in returning the base or dictionary form of a word known as the lemma.

```python
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()
e_words = ["wait", "waiting", "waited", "waits"]
for w in e_words:
	print("Lemma for {} is {}".format(w, wordnet_lemmatizer.lemmatize(w)))
```

```python
def lemmatize(text):
	return wordnet_lemmatizer.lemmatize(text)
```

**Removing Stopwords**

```python
nltk.download('stopwords')
stopwords_list = stopwords.words('english')
stopwords_list
```

```python
def removestopwords(text):
	clean = " ".join([word for word in text.split() if word not in stopwords_list])
	return clean

#demo
removestopwords("this weather")
```

**Defining a function to preprocess the text using lambda functions**

```python
df_train['class_label'] = df_train['class_label'].replace(['Yes','No'],[1,0])
df_dev['class_label'] = df_dev['class_label'].replace(['Yes','No'],[1,0])
def preprocess(df):
	df['Text']= df['Text'].apply(lambda x:tolower(x))
	df['Text']= df['Text'].apply(lambda x:removepunctuation(x))
	df['Text']= df['Text'].apply(lambda x:removestopwords(x))
	df['Text']=df['Text'].apply(lambda x:lemmatize(x))
	return df
```

```python
#Converting the Text to a string and class_label to integer
df_train.Text=df_train.Text.astype(str)
df_train.class_label=df_train.class_label.astype(int)
#Applying the function
train = preprocess(df_train)
train_data=train[["Text","class_label"]]
train_data
```

```python
#Converting the Text to a string and class_label to integer
df_dev.Text=df_dev.Text.astype(str)
df_dev.class_label=df_dev.class_label.astype(int)
#Applying the function
dev = preprocess(df_dev)
dev_data=dev[["Text","class_label"]]
dev_data
```

```python
#Converting the Text to a string and class_label to integer
df_test.Text=df_test.Text.astype(str)
#df_test.class_label=df_test.class_label.astype(int)
#Applying the function
test = preprocess(df_test)
test_data=test[["Text"]]
test_data
# not needed
```

# APPLYING TRANSFORMER MODELS

```python
def find_accuracy(bert,df):
    result, model_outputs, wrong_predictions = bert.eval_model(df)

    predictions, raw_outputs = bert.predict(df['Text'].tolist())
    print("predictions:")
    print(predictions)

    bads = {}
    for pred in wrong_predictions:
        if pred.label in bads:
            bads[pred.label] += 1
        else:
            bads[pred.label] = 1
    print("wrong predictions:")
    wrong = len(wrong_predictions)
    total = len(df)
    print(str(wrong) + ' wrong out of ' + str(total))
    print(bads)
    accuracy = (total-wrong)/total*100
    print("Accuracy: "+str(accuracy)+"%")

    print("Classification report: ")
    from sklearn.metrics import classification_report
    print(classification_report(df.class_label, predictions))
```

```python
def predict_result(bert,df):
    predictions, raw_outputs = bert.predict(df['Text'].tolist())

    return predictions
```

# TRADITIONAL METHODS

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

texts_train = train_data["Text"]
labels_train = train_data["class_label"]
texts_test = test_data["Text"]
labels_test = test_data["class_label"]

# Convert the texts into feature vectors
vectorizer = CountVectorizer()
X_train = vectorizer.fit_transform(texts_train)
X_test = vectorizer.transform(texts_test)
```

```python
model = MultinomialNB()
model.fit(X_train, labels_train)
```

```python
labels_pred = model.predict(X_test)
```

```python
report = classification_report(labels_test, labels_pred)
print(report)
```

```python
model = SVC(kernel='linear')
model.fit(X_train, labels_train)
```

```python
labels_pred = model.predict(X_test)
```

```python
report = classification_report(labels_test, labels_pred)
print(report)
```

```python
model = LogisticRegression()
model.fit(X_train, labels_train)
```

```python
labels_pred = model.predict(X_test)
```

```python
report = classification_report(labels_test, labels_pred)
print(report)
```
